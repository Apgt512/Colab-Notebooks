**Inversa de una matriz**: Es análogo a la inversa de un número.  
N es la inversa de la matriz M si:  
$M\cdot N=I$

**Determinante**: Propiedad de las matrices cuadradas. Nos indica si la matriz tiene inversa. Para una matriz de 2x2:  
$|M|=\begin{bmatrix}
a & b\\
c & d
\end{bmatrix}=ad-bc$

```python
M = np.random.rand(4).reshape(2,2)
print(M)

# Calcular el determinante
print('\nDeterminante:', np.linalg.det(M))

# Calcular la inversa de una matriz
print('\nInversa:\n', np.linalg.inv(M))
```

```python
M_sin_inversa = np.ones((2,2))*2
print(np.linalg.det(M_sin_inversa))
np.linalg.inv(M_sin_inversa)
```

**Rango**: El rango de una matriz A de m x n es el número de columnas o filas linealmente independientes.

```python
print('Rango M:', np.linalg.matrix_rank(M))
print('Rango M_sin_inv:', np.linalg.matrix_rank(M_sin_inversa))
```

Para dos vectores $x$ e $y$ y dos escalares $a$ y $b$, $F$ es una **transformación lineal** si:  
$F(ax+by)=aF(x)+bF(y)$

```python
x = np.array([1,2])
y = np.array([3,4])

a = 2
b = 3
# Saquemos el valor negativo de nuestro vector
def f(vector):
    return -1*vector

print(f(2*x+3*y))
print(a*(f(x))+b*f(y))
```

### Sistemas de ecuaciones lineales ###
Una ecuación lineal es una igualdad que tiene la forma:  
$\sum{a_i x_i}=y$  
Un sistema de ecuaciones lineales es un set de ecuaciones lineales que comparten las mismas variables. Por ejemplo:  
$a_1x_1 + a_2x_2 = y_1$  
$a_3x_1 + a_4x_2 = y_2$  

La forma matricial de un sistema de ecuaciones lineales se expresa como $Ax=y$  

- Caso 1: No hay solución.
- Caso 2: Existe una única solución.
- Caso 3: Existen infinitas soluciones.  

Si tenemos un sistema como el anterior $Ax=y$ donde $A$ es una matriz de $m \times n$ significa que tenemos un sistema con $m$ ecuaciones y $n$ incógnitas.
